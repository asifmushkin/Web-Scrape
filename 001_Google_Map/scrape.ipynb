### Importing modules

```python
from dotenv import load_dotenv
import pandas as pd
import os
```

### importing serp api key from env file

```python
serp_api_key = os.getenv("SERP_API_KEY")
```

#### source code

```python
from serpapi import GoogleSearch

results_list = []
for i in range(3):    
  params = {
    "engine": "google_maps",
    "q": "hotels",
    "ll": "@25.0657,55.17128,15.1z",
    "type": "search",
    "api_key": serp_api_key,
    "start":i
  }

  search = GoogleSearch(params)
  results = search.get_dict()
  local_results = results["local_results"]
  results_list.extend(local_results)
```

### display results

```python
results_list
```

### creating data frame from json

```python
df_results = pd.DataFrame(results_list)
```

#### displaying  data frame output

```python
df_results
```

#### Converting data frame into excel format

```python
df_results.to_excel("results.xlsx",index=True)
```
