### Importing modules

```python
from dotenv import load_dotenv
import pandas as pd
import os
```

### importing serp api key from env file

```python
serp_api_key = os.getenv("SERP_API_KEY")
```

#### source code
scrapping a specific no of pages using for loop

```python
from serpapi import GoogleSearch

results_list = []
for i in range(3):    
  params = {
    "engine": "google_maps",
    "q": "hotels",
    "ll": "@25.0657,55.17128,15.1z",
    "type": "search",
    "api_key": serp_api_key,
    "start":i
  }

  search = GoogleSearch(params)
  results = search.get_dict()
  local_results = results["local_results"]
  results_list.extend(local_results)
```

### display results

```python
results_list
```

### creating data frame from json

```python
df_results = pd.DataFrame(results_list)
```

#### displaying  data frame output

```python
df_results
```

#### Converting data frame into excel format

```python
df_results.to_excel("results.xlsx",index=True)
```

#### Scrapping all available pages using while loop

```python
from serpapi import GoogleSearch

results_list = []
page_no =0
while True:    
  params = {
    "engine": "google_maps",
    "q": "hotels",
    "ll": "@25.0657,55.17128,15.1z",
    "type": "search",
    "api_key": serp_api_key,
    "start":page_no
  }

  search = GoogleSearch(params)
  results = search.get_dict()
  if "local_results" in results:

    local_result = results["local_results"]
    results_list.extend(local_result)
    page_no += 1

  else:
    print("All available pages are being scrapped and no more pages exist.")
    break
```

### Autocompeletion API

```python
from serpapi import GoogleSearch

params = {
  "engine": "google_maps_autocomplete",
  "q": "cafe",
  "hl": "en",
  "gl": "ae",
  "ll":"@23.424076,53.847818,15z",
  "api_key": serp_api_key
}

search = GoogleSearch(params)
results = search.get_dict()
suggestions = results["suggestions"]
```
